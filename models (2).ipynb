```python
import pandas as pd
```

```python
df = pd.read_csv('Car Dataset Processed.csv')
```

```python
df.columns
```

```python
df['insurance_validity'].value_counts()
```

```python
d1 = {'Comprehensive':0,'Third Party insurance':1,
                              'Zero Dep':2,'Not Available':3,'Third Party':1}
df['insurance_validity'] = df['insurance_validity'].map(d1)
```

```python
df['insurance_validity'].value_counts()
```

```python
d2 = {'Petrol':0,'Diesel':1,'CNG':2}
df['fuel_type'] = df['fuel_type'].map(d2)
df['fuel_type'].value_counts()
```

```python
d3 = {'Manual':0,'Automatic':1}
df['transmission'] = df['transmission'].map(d3)
df['transmission'].value_counts()
```

```python
d4 = {'First Owner':1,'Second Owner':2,'Third Owner':3,'Fifth Owner':5,'Fourth Owner':4}
df['ownsership'] = df['ownsership'].map(d4)
```

```python
df['ownsership'].value_counts()
```

```python
# df['registration_year']
```

```python
X = df[['insurance_validity','fuel_type','kms_driven','ownsership','transmission']]
y = df['price(in lakhs)']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
```

```python
models = {'LinReg':LinearRegression(),
          'KNN-3':KNeighborsRegressor(n_neighbors=3),
          'SVM-Poly':SVR(kernel='poly')}
for name,model in models.items():
    model.fit(X,y)
    sc= model.score(X,y)
    print(name,sc)
```

```python
final_model = models['KNN-3']
```

```python
d1={'Comprehensive': 0,'Third Party insurance': 1,'Zero Dep': 2,'Not Available': 3,'Third Party': 1}
d2={'Petrol': 0, 'Diesel': 1, 'CNG': 2}
d3={'Manual': 0, 'Automatic': 1}
d4={'First Owner': 1,'Second Owner': 2,'Third Owner': 3,'Fifth Owner': 5,'Fourth Owner': 4}
```

```python
insurance_validity = input('Insurance validity:')
fuel_type = input('Fuel Type:')
kms_driven = input('KMs Drive:')
ownsership = input('Ownership:')
transmission = input('Transmission Type:')
insurance_validity = d1[insurance_validity]
fuel_type = d2[fuel_type]
transmission = d3[transmission]
ownsership = d4[ownsership]
kms_driven = int(kms_driven)
test = [[insurance_validity,fuel_type,kms_driven,ownsership,transmission]]
yp = int(final_model.predict(test)[0])
print('Predicted Car Price is {} Rs.'.format(yp))
```

```python
models = {
    'LinReg': LinearRegression(),
    'KNN-3': KNeighborsRegressor(n_neighbors=3),
    'SVM-Poly': SVR(kernel='poly')
}

for name, model in models.items():
    model.fit(X, y)
    sc = model.score(X, y)
    print(name, sc)

final_model = models['KNN-3']   
```

```python
import pickle

# Save the trained KNN model
pickle.dump(final_model, open("car_price_model.pkl", "wb"))

# Save your encoding dictionaries too
pickle.dump((d1, d2, d3, d4), open("encoders.pkl", "wb"))
```
